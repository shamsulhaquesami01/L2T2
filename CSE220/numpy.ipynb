```python
class Employee:
    class_name = 'Employee'

    # double underscore -> dunder
    def __init__(self, id, name):
        self.id = id
        self.name = name

    def __str__(self):
        return f"{self.id} with name {self.name}"
    
    def __repr__(self):
        return f"{self.name} has the id {self.id}"

employee = Employee(2596, "Kirk")
print(employee)
employee
```

```python
a = np.array([[1, 2, 3, 4], [5, 6, 4, 8], [9, 10, 11, 12]])

# 1. Evaluate the condition array
mask = a < 5 
# This generates a boolean matrix of the same shape filled with True/False values

# 2. Apply mask as index coordinates
print(a[mask]) 
# Output: [1, 2, 3, 4] [cite: 214]
```

```python
filtered = a[(a > 2) & (a < 11)] 
# Output: [ 3,  4,  5,  6,  7,  8,  9, 10] [cite: 218]
print(filtered)
```
